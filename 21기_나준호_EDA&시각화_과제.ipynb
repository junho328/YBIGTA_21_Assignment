### 2) 결측치 확인

- #### <font color='orange'> Q2. 각 컬럼의 결측치를 확인하고, 결측치가 많은 순서대로 결측치 비율과 함께 표로 나타내보세요.
<font color='orange'> (결측치 0개인 컬럼은 나타내지 마세요)

```python
# Q2
df.info()
```

```python
missingnum = df.isnull().sum()
missingnum = missingnum[missingnum>0]
q2df = pd.DataFrame(missingnum, columns=['missing_num'])
q2df = q2df.sort_values(by='missing_num', ascending = False)
q2df['missing_%'] = round(q2df['missing_num']/len(df)*100,1)
q2df
```

- #### <font color='orange'> Q3. 결측치가 있는 컬럼과 결측치 비율을 보기쉽게 그래프로 나타내 보세요.

```python
# Q3
plt.figure(figsize=(10,5))
sns.barplot(x=q2df.index, y=q2df['missing_%'].values)
plt.xticks(rotation = 45, fontsize = 10)
plt.show()
```

- #### <font color='orange'> Q4. 결측치가 적은 beds 컬럼을 좀 더 자세히 살펴보겠습니다. 
  #### <font color='orange'> 'beds' 컬럼이 결측치인 데이터의 상위 5개를 출력해주세요.

```python
# Q4
df[df['beds'].isnull()].head()
```

### 3) 범주형 변수 확인
<font color='gray'> - 편의를 위해 결측치가 있는 컬럼을 모두 제거하고 진행하겠습니다.

```python
drop_col = ['jurisdiction_names', 'license', 'square_feet', 'monthly_price',
       'weekly_price', 'host_response_rate', 'host_response_time',
       'security_deposit', 'host_acceptance_rate', 'review_scores_location',
       'review_scores_value', 'review_scores_checkin',
       'review_scores_accuracy', 'review_scores_communication',
       'review_scores_cleanliness', 'review_scores_rating',
       'reviews_per_month', 'beds']
```

```python
df = df.drop(columns=drop_col)
```

- #### 각 컬럼별 변수 타입 확인

```python
df.info()
```

- #### <font color='orange'> Q5. Room Type 에는 어떤 것들이 있는지 살펴보겠습니다. 
#### <font color='orange'> room_type의 종류별 갯수에 대한 그래프를 그려주세요.

```python
# Q5
sns.countplot(x='room_type',data=df)
plt.show()
```

- #### <font color='orange'> Q6. 위 그래프의 세로축이 갯수가 아닌 비율을 나타내도록 재구성해보세요.

```python
sns.barplot(x=df.room_type.value_counts().index, y=df.room_type.value_counts()/len(df))
plt.show()
```

- #### <font color='orange'> Q7. 각 지역별로 room_type의 비율이 어땠는 지 알고 싶습니다.
#### <font color='orange'> 아래 그래프와 같이 각 지역별로 room_type의 갯수가 어땠는지 보여주세요.    

```python
sns.set() #plt를 sns 설정으로 유지
plt.figure(figsize=(10,5)) #피겨 사이즈 설정
sns.countplot(x = 'neighborhood', data = df, hue = 'room_type')
plt.show()
```

- #### <font color='orange'> Q8. 각 지역별로 가격의 분포가 어떤지 알고 싶습니다.
#### <font color='orange'> 아래 그래프와 같이 각 지역별로 가격의 분포를 보여주세요.

```python
plt.figure(figsize=(10,5))
sns.stripplot(x = 'neighborhood', y = 'price', data = df)
plt.show()
```

```python
# Q9

1. 맨하튼,브루클린,퀸즈에는 이상치가 존재한다.
2. 특히 맨하튼의 경우에는 outlier, 극도로 비싼 가격의 집들이 많다.
3. 그에 비해 staten island,bronx는 비교적으로 outlier 없이 가격이 균일하게 분포되어있다.
```
