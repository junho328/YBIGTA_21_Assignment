## 2. 지표변수

```python
old = []
for i in df['Age']:
    if (1 <= i <= 19):
        old.append('young')
    elif (20 <= i <= 50):
        old.append('middle')
    else:
        old.append('old')
        
df['old'] = old
df.head()
```

```python
## 3. Feature Split
df['date'] = '1914-04-14'
df
```

```python
year = []
month = []
day = []

for j in df['date'].str.split('-'):
    year.append(j[0])
    month.append(j[1])
    day.append(j[2])
df['year'] = year
df['month'] = month
df['day'] = day
```

```python
df
```

## 4. 스케일링

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
mm_scaler = MinMaxScaler()
sd_scaler = StandardScaler()
```

```python
mm_scaler = mm_scaler.fit_transform(df['Age'].values.reshape(-1,1))
df['mm_scaled_age'] = mm_scaler

sd_scaler = sd_scaler.fit_transform(df['Age'].values.reshape(-1,1))
df['sd_scaled_age'] = sd_scaler
```
