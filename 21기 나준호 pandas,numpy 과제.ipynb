```python
# df_x와 df_y를 inner join시켜라 (quality를 기준으로)
pd.merge(df_x,df_y, how='inner',on='quality')
```

```python
# df_x에서 df_y를 left join시켜라 (quality를 기준으로)
pd.merge(df_x,df_y, how='left', on='quality')
```

# 과제 2 

```python
df = pd.read_csv('./deliver.csv')
df.head()
```

```python
# 1. 날짜' 컬럼을 보고, '월' 컬럼과 '계절' 컬럼을 새로 추가하라. (계절 컬럼 : 3-5월 : 봄, 6-8월 : 여름, 9-11월: 가을, 12-2월 : 겨울)
import datetime

def month(date_string):
    date = datetime.datetime.strptime(date_string, "%Y-%m-%d")
    return date.month

def season(date_string):
    date = datetime.datetime.strptime(date_string, "%Y-%m-%d")
    if 3 <= date.month <= 5:
        return '봄'
    if 6 <= date.month <= 8:
        return '여름'
    if 9 <= date.month <= 11:
        return '가을'
    if 1 <= date.month <= 2 or date.month == 12:
        return '겨울'
    
df['월'] = df['날짜'].apply(month)
df['계절'] = df['날짜'].apply(season)
df.head()
```

```python
# 2. '미세먼지(㎍/㎥)' 컬럼에서 null값의 개수를 확인하고, null값을 전체 평균으로 대체하라. (전체 평균은 null값이 아닌 값들의 평균으로 계산된다.)
df['미세먼지(㎍/㎥)'].isnull().sum()
```

```python
df['미세먼지(㎍/㎥)'].fillna(df['미세먼지(㎍/㎥)'].mean(),inplace=True)
df.loc[:10]
```
